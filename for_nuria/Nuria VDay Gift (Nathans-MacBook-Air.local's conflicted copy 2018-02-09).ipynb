```python
import numpy as np
import matplotlib as mpl
mpl.use('nbagg')
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
from skimage import feature
from skimage.color import rgb2gray
```

```python
rgb_image = plt.imread('n&n.jpg')
rgb_image = np.array(rgb_image, dtype=np.float64)/255
img_shape = rgb_image.shape
```

```python
plt.imshow(rgb_image)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(4).fit(np.reshape(rgb_image, (-1,3)))
```

```python
quantized_image = np.ndarray(img_shape)
rows, cols, channels = img_shape
for row in range(rows):
    for col in range(cols):
        quantized_image[row,col] = kmeans.cluster_centers_[kmeans.labels_[row*cols+col]]
```

```python
plt.imshow(quantized_image); plt.show()
```

```python
bw_image = rgb2gray(rgb_image)
plt.imshow(bw_image, cmap='gray');plt.show()
```

```python
edges = feature.canny(bw_image, sigma=0.9, low_threshold=0.1, high_threshold=0.2)
plt.imshow(edges, cmap='gray'); plt.show()
```
