```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# True variance parameter of population
true_mean = 0.0
true_var = 2.6341

# Sample for calculating variance estimate
sample_size = 100
sample = np.random.normal(true_mean, true_var**0.5, sample_size)
```

```python
print('Mean Estimate Error: ', sample.mean() - true_mean)
print('Variance Estimate Error: ', sample.var() - true_var)
```

```python
num_resamples = 1000
resample_sizes = np.arange(2,99)
bootstrap_vars = []
bootstrap_means = []
for resample_size in resample_sizes:
    bootstrap_var = 0
    bootstrap_mean = 0
    for i in range(num_resamples):
        resample = np.random.choice(sample, resample_size, replace=True)
        bootstrap_var += resample.var()/num_resamples
        bootstrap_mean += resample.mean()/num_resamples
    bootstrap_vars.append(bootstrap_var)
    bootstrap_means.append(bootstrap_mean)
```

```python
ax = plt.subplot(111)
ax.plot(np.array(bootstrap_vars)-true_var, label='Var')
ax.plot(np.array(bootstrap_means)-true_mean, label='Mean')
ax.set_ylabel('Estimate Error')
ax.set_xlabel('Resample Size')
ax.legend()
plt.show()
```
