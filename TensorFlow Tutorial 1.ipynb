```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter((1,2,2.5), (2,1,2)); plt.xlim((0,3)); plt.ylim((0,3));
```

```python
A = tf.constant(((1,1),(2,1),(2.5,1)))
b = tf.constant((2.,1,2), shape=(3,1))
x = tf.get_variable('parameters', initializer=tf.constant((1.0,1.0), shape=(2,1)), dtype=tf.float32)
```

```python
cost = tf.reduce_sum(tf.square(tf.matmul(A,x)-b))
opt = tf.train.AdamOptimizer().minimize(cost)
```

```python
initializer = tf.global_variables_initializer()
sess = tf.Session()
sess.run(initializer)
```

```python
costs =[]
for iterations in range(int(5e3)):
    _,c = sess.run((opt,cost))
    costs.append(c)
```

```python
params = sess.run(x)
```

```python
import numpy as np
xvals = np.linspace(0,3,1000)
yvals = params[0]*xvals + params[1]
y_init_vals = xvals + 1
```

```python
plt.scatter((1,2,2.5),(2,1,2)); plt.plot(xvals, y_init_vals); plt.xlim((0,3)); plt.ylim((0,3));
```

```python
plt.scatter((1,2,2.5),(2,1,2)); plt.plot(xvals, yvals); plt.xlim((0,3)); plt.ylim((0,3));
```
