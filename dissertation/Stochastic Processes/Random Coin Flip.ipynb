We simulate a coin flipping game where each heads results in earning 1 dollar and each tails results in losing one dollar. How does our profit behave over time?

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate(num_flips):
    return np.cumsum(2*np.random.binomial(1,0.5,num_flips)-1)
```

```python
n=100000
x_vals = np.linspace(0,n,n)
plt.plot(simulate(n))
plt.plot(np.sqrt(x_vals))
plt.plot(-np.sqrt(x_vals))
```
