```python
import numpy as np

def f(x):
    return x*x

def h(x):
    return x

import matplotlib.pyplot as plt

x = np.linspace(0.001,1,200)
plt.plot(x,f(x),label='$f(X)$')
plt.plot(x,h(x),label='$h(x)$')
plt.plot(x,h(x)*np.log(f(x)),label='$h(x)\log{f(x)}$')
plt.plot(x,np.log(h(x)),label='$f(x)\log{h(x)}$')

plt.legend()
plt.show()
```

```python
def ji(x,a):
    return (1-0.1*(1/x)**a)/a
def di(x):
    return np.log(x/2)+1
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(1e-5,2,1000)
plt.plot(x,ji(x,-0.01),label='ji')
plt.plot(x,di(x),label='di')
plt.legend()
plt.show()
```
