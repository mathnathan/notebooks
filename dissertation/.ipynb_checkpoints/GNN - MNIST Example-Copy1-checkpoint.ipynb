```python
import matplotlib.pyplot as plt
%matplotlib inline
from ipywidgets import interactive
import numpy as np
import matplotlib.patches as patches
from Brain import Neuron, GMM, Net
from collections import defaultdict

#p = GMM([0.2,0.8], np.array([[0.24,0.01],[0.78,0.01]]))
```

```python
import torch
import torchvision.datasets as datasets
mnist_trainset = datasets.MNIST(root='./data', train=True, download=True, transform=None)
```

```python
network = Net(28,28,1000)
weights = np.random.rand(1000)
one_neuron = Neuron([1000,1],weights,0.01,0.00001)
```

```python
one_guesses = []
for i,(im,label) in enumerate(mnist_trainset):
    if label==1:
        feedback=10
    else:
        feedback=-0.1
    y = network(np.array(im),feedback)
    is_it_a_one = one_neuron(y,feedback)
    if label==1:
        one_guesses.append(is_it_a_one)
        
    if i%100 == 0:
        print('one_neuron.bias = ', one_neuron.bias)
        print('mean(one_neuron.weights) = ', one_neuron.weights.mean())
        biases = [sum([n.bias for n in b]) for b in network.neurons.values()]
        print('max bias = ', max(biases))
        print('min bias = ', min(biases))
        plt.plot(one_guesses)
        plt.show()
```
